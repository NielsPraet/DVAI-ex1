```python
import pandas as pd
df = pd.read_csv("belgian_municipalities.csv")
```

```python
df.head()
```

```python
columns = ["municipality_NL", "arrondissement_NL", "province_NL", "region_NL", "inhabitants"]

# Clean up data
df.loc[df['region_NL'] == "Brussels Hoofdstedelijk Gewest", "province_NL"] = "Brussel"
df.loc[df['region_NL'] == 'Henegouwen', 'region_NL'] = "Waals Gewest"
df.loc[df['region_NL'] == 'Région wallonne', 'region_NL'] = "Waals Gewest"

df = df[columns]
df.head()
```

```python
# use plotly for sunburst
import plotly.express as px

fig = px.sunburst(df, path=['region_NL', 'province_NL', 'arrondissement_NL'], values='inhabitants')
fig.update_layout(width=1000, height=1000)
fig.show()
```

```python
fig = px.sunburst(df, path=['region_NL', 'province_NL'], values='inhabitants')
fig.show()
```

```python
fig = px.sunburst(df, path=['province_NL', 'arrondissement_NL', 'municipality_NL'], values='inhabitants')
fig.show()
```

```python
fig = px.sunburst(df[df['region_NL'] == 'Vlaams Gewest'], path=['province_NL', 'arrondissement_NL', 'municipality_NL'], values='inhabitants')
fig.show()
```

```python
fig = px.sunburst(df[df['region_NL'] == 'Waals Gewest'], path=['province_NL', 'arrondissement_NL', 'municipality_NL'], values='inhabitants')
fig.show()
```
