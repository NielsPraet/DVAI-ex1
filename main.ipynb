```python
import pandas as pd
df = pd.read_csv("belgian_municipalities.csv")
```

```python
df.head()
```

```python
columns = ["municipality_NL", "arrondissement_NL", "province_NL", "region_NL", "inhabitants"]

df = df[columns]
df.head()
```

```python
# use plotly for sunburst
import plotly.express as px

fig = px.sunburst(df, path=['region_NL', 'province_NL', 'arrondissement_NL', 'municipality_NL'], values='inhabitants')
fig.show()
```
