```python
import pandas as pd
df = pd.read_csv("belgian_municipalities.csv")
```

```python
df.head()
```

```python
columns = ["municipality_NL", "arrondissement_NL", "province_NL", "region_NL", "inhabitants"]

# Fill empty label with more meaningful name
df.loc[df['region_NL'] == "Brussels Hoofdstedelijk Gewest", "province_NL"] = "Brussel"

df = df[columns]
df.head()
```

```python
# use plotly for sunburst
import plotly.express as px

fig = px.sunburst(df, path=['region_NL', 'province_NL', 'arrondissement_NL', 'municipality_NL'], values='inhabitants')
fig.show()
```

```python


fig = px.sunburst(df, path=['region_NL', 'province_NL'], values='inhabitants')
fig.show()
```
